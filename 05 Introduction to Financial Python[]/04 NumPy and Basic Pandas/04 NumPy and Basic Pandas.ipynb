```python
import pandas as pd
```

```python
price = [143.73, 145.83, 143.68, 144.02, 143.5, 142.62]
s = pd.Series(price)
s
```

```python
s = pd.Series(price,index = ['a','b','c','d','e','f'])
s
```

```python
s.index = [6,5,4,3,2,1]
s
```

```python
print(s[1:])
print(s[:-2])
```

```python
print(s[4])
s[4] = 0
print(s)
```

```python
s = pd.Series(price, name = 'Apple Price List')
print(s)
print(s.name)
```

```python
time_index = pd.date_range('2017-01-01',periods = len(s),freq = 'D')
print(time_index)
s.index = time_index
print(s)
```

```python
s.index = [6,5,4,3,2,1]
print(s)
print(s[1])
```

```python
print(s.iloc[1])
```

```python
s.index = time_index
print(s['2017-01-03'])
```

```python
print(s['2017-01-02':'2017-01-05'])
```
